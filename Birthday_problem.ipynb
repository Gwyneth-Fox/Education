```python
import pandas as pd
import random
import matplotlib.pyplot as plt

def choose_date():
    return random.randint(1,365)

data = []

for i in range(10000):
    count = 0 
    chosen_dates = set()
    new_date = choose_date()
    not_done = True
    chosen_dates.add(new_date)
    while not_done:
        new_date = choose_date()
        if new_date in chosen_dates:
            not_done = False
        else:
            chosen_dates.add(new_date)
    
    d = {"Number of days":len(chosen_dates)}
    data.append(d)

df = pd.DataFrame(data)
df





```
